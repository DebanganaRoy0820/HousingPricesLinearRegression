```python
y = data['Y house price of unit area']
X = data.drop('Y house price of unit area',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

**Splitting into training and testing data**

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

**Create an instance of the LinearRegression() model named lm.**

```python
lm.fit(X_train,y_train)
```

**Train/fit lm on the training data.**

```python
lm.coef_
```

**Print out the coefficients of the model**

```python
X_train.columns
```

```python
pd.DataFrame(lm.coef_,X.columns,columns=['Coeffs'])
```

```python
predictions = lm.predict(X_test)
```

**Predicting Test Data**

```python
predictions
```

```python
fig = plt.scatter(data,y=y_test,x=predictions)
fig.show()
```

**Create a scatterplot of the real test values versus the predicted values.**

```python
from sklearn import metrics
print('MAE',metrics.mean_absolute_error(y_test,predictions))
print('MSE',metrics.mean_squared_error(y_test,predictions))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

**Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.**

```python
sns.distplot((y_test-predictions),bins=50)
```
